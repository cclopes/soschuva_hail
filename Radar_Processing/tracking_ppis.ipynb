### Funções
#### Descrição delas em *funcoes_descricao*

```python
def plota_ppi(arquivo, elevacao, lon_lims = [-47.3, -46.8], lat_lims = [-23.1, -22.7], 
             variavel = 'corrected_reflectivity', var_min = 10, var_max = 70):
    
    display = pyart.graph.RadarMapDisplay(arquivo)
    fig = plt.figure(figsize = [4,3])
    display.plot_ppi_map(variavel, sweep = elevacao, shapefile = "shapefiles/sao_paulo", vmin = var_min, vmax = var_max, 
                         min_lat = lat_lims[0], max_lat = lat_lims[1], min_lon = lon_lims[0], max_lon = lon_lims[1], 
                         lat_lines = np.arange(lat_lims[0], lat_lims[1], .5), 
                         lon_lines = np.arange(lon_lims[0], lon_lims[1], .5))
    display.plot_point(lat=-22.81405, lon=-47.05641, symbol = 'k.', label_text='hailpad')
```

```python
for i in caixas:
    srfile = pyart.aux_io.read_gamic(i[2])
    
    #-- Plotando
    display = pyart.graph.RadarMapDisplay(srfile)
    fig = plt.figure(figsize=(6,5))
    xlim, ylim = [-47.5,-46.5], [-23.35,-22.5]
    centro_sistema = (np.mean(i[4]), np.mean(i[5]))
    display.plot_ppi_map('corrected_reflectivity', sweep=1, vmin=10, vmax=70, shapefile="shapefiles/sao_paulo",
                         mask_outside=True, min_lat = ylim[0], max_lat = ylim[1], min_lon = xlim[0], max_lon = xlim[1], 
                         lat_lines=np.arange(ylim[0], ylim[1], .25), lon_lines=np.arange(xlim[0], xlim[1], .25))
    display.plot_point(lat=i[6][0], lon=i[6][1], symbol = 'ks', #-- Hailpad
                       markersize=10, alpha=0.9, markerfacecolor='snow')
    display.plot_point(lat=centro_sistema[1], lon=centro_sistema[0], symbol = 'k*', #-- Centro da caixa do sistema
                       markersize=15, alpha=0.9, markerfacecolor='snow')
    plt.savefig('figuras/ppis/classificacao/sr_h_dBZ_d'+str(i[1])+'.png', bbox_inches='tight')
    
    #-- Selecionando seção transversal
    sr_cs = pyart.util.cross_section_ppi(srfile, [i[0]])
    #-- Plotando
    display = pyart.graph.RadarDisplay(sr_cs)
    #--- Refletividade
    f = plt.figure(figsize = [6,3])
    ax = f.add_subplot(111)
    display.plot('corrected_reflectivity', vmin = 10, vmax = 70,
                 title = 'Azimuth = '+str(i[0])+' Deg. '+sr_cs.time['units'][14:]+'\nCorrected Reflectivity')
    pnt = ax.scatter(x=i[1], y=1, marker='s', c='snow', alpha=0.9, edgecolors='black') #-- Hailpad
    plt.xlim(i[3])
    plt.ylim((0,18))
    plt.savefig('figuras/ppis/classificacao/sr_v_dBZ_d'+str(i[1])+'.png', bbox_inches='tight')
```
