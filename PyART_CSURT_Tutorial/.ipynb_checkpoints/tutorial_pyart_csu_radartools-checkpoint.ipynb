## Plotando os dados de radar

Dentro do Py-ART, existem várias funções para plotar campos horizontais e cortes verticais. Elas são chamadas a partir de displays que plotam os dados brutos de radar ou já em uma grade (todos os displays e suas respectivas funções estão descritas na  [Documentação do Py-ART](http://arm-doe.github.io/pyart/dev/user_reference/graph.html).  
Nestes casos, serão plotados:
- Campos horizontais de PPI de uma dada variável em uma dada elevação
    - Display `RadarMapDisplay` para mapear os dados em uma grade cartesiana e adicionar o shapefile
- Cortes verticais de uma dada variável em um determinado azimute
    - Display `RadarDisplay`  

Todas essas funções foram derivadas do pacote Matplotlib. Assim, podem ser usados argumentos adicionais para customizar os gráficos de acordo com a função do Matplotlib que é compatível com a função do Py-ART.

### Campos horizontais

Como as variáveis `cth` e `sr` tem o mesmo formato, elas podem ser plotadas com a mesma função!  
Ela se chama `plot_ppi_map` e precisa dos seguintes argumentos:
- `field`: campo da variável a ser plotada
- `sweep`: elevação a ser plotada (**lembre-se: o Python começa a contar do zero!**)  

Outros argumentos importantes são:
- `min_lat`, `max_lat`, `min_lon`, `max_lon`: limites do gráfico
- `vmin`, `vmax`: limites do campo
- `lat_lines`, `lon_lines`: vetores definindo as linhas de grade - **o que faz plotar os valores nos eixos horizontal e vertical!**
- `shapefile`: localização dos arquivos de shapefile (caso você queira maior detalhamento do que as fronteiras dos estados)  
_Este argumento é descrito como "experimental", ou seja, pode não funcionar. Pelo que eu entendi, o melhor display para se usar com shapefile é o `RadarMapDisplayCartopy`, que conversa com o pacote Cartopy (próxima geração do Basemap), mas como as poucas experiências que tive com o Cartopy mostraram vários problemas de compatibilidade com o próprio Basemap, ainda não arrisquei usá-lo. Uma limitação de se usar o `RadarMapDisplay` com shapefile é que não tem como modificar esse shapefile, mostrando apenas parte dele, por exemplo._

Como essas funções foram criadas a partir do Matplotlib, sua estrutura de "montagem do plot" precisa ser usada.  
Essa estrutura chama várias funções dentro de objetos dentro de outros objetos, todos dentro de um mesmo pacote. Assim, para facilitar o entendimento e diminuir o tamanho do código, é importante definir outras variáveis de "meio de caminho".  
Note que, ao carregar os pacotes necessários, isso já foi feito com o pacote __matplotlib.pyplot__, que foi chamado de __plt__. Assim, em vez de chamar o `matplotlib.pyplot.figure`, você pode chamar o `plt.figure`. Essa nomenclatura (matplotlib.pyplot = plt) é considerada padrão no Python, assim como chamar o numpy de np, entre outros.   
  
Para plotar:
- Chame o display da variável a ser plotada
- Crie a figura com o tamanho e outras configurações desejadas
- Plote
- Defina outros parâmetros relacionados ao plot (opcional, comum para funções nativas do `plt`)

## Calculando massas de água e gelo

O pacote CSU_RadarTools fornece diversas ferramentas para calcular outras grandezas a partir das variáveis polarimétricas do radar. Aqui será mostrada o cálculo das massas de água e gelo.  

**Para ver outras ferramentas, veja o notebook [CSU_RadarTools Demo](https://github.com/CSU-Radarmet/CSU_RadarTools/blob/master/notebooks/CSU_RadarTools_Demo.ipynb).**  
  
O cálculo das massas de água e gelo é feito com a função `calc_liquid_ice_mass`, que precisa das seguintes variáveis:
- Refletividade ou Refletividade Corrigida
- Refletividade Diferencial
- Temperatura (opcional, obtida a partir de radiossondagem)  

__Como essa função precisa de uma variável polarimétrica, ela só será aplicada aos dados da FCTH.__

Antes de efetivamente usar a função, é preciso criar outras duas funções:
- `interpolate_sounding_to_radar`: Interpola os dados da radiossondagem para o mesmo formato dos dados de radar para poder calcular as massas
- `add_field_to_radar_object`: Adiciona um campo à variável de radar para plotar as massas da mesma forma que os outros campos do radar  

Essas funções precisam de outras funções que também serão criadas:
- `get_z_from_radar`: calcula altura correspondente à elevação do radar
- `check_sounding_for_montonic`: checa se a radiossondagem lida tem um perfil monotônico (sem saltos); se não, corrige
- `radar_coords_to_cart`: converte as coordenadas de radar para cartesianas

### `interpolate_sounding_to_radar`: interpolando dados de radiossondagens aos de radar

```python
def interpolate_sounding_to_radar(sounding, radar):
    radar_z = get_z_from_radar(radar)
    radar_T = None
    snd_T, snd_z = check_sounding_for_montonic(sounding)
    shape = np.shape(radar_z)
    rad_z1d = radar_z.ravel()
    rad_T1d = np.interp(rad_z1d, snd_z, snd_T)
    return np.reshape(rad_T1d, shape), radar_z
```

### `get_z_from_radar`: Calculando altura do radar correspondente às elevações

```python
def get_z_from_radar(radar):
    azimuth_1D = radar.azimuth['data']
    elevation_1D = radar.elevation['data']
    srange_1D = radar.range['data']
    sr_2d, az_2d = np.meshgrid(srange_1D, azimuth_1D)
    el_2d = np.meshgrid(srange_1D, elevation_1D)[1]
    xx, yy, zz = radar_coords_to_cart(sr_2d/1000.0, az_2d, el_2d)
    return zz + radar.altitude['data']
```

### `check_sounding_for_montonic`: Forçando os dados de sondagem a serem monotônicos

```python
def check_sounding_for_montonic(sounding):
    snd_T = sounding.soundingdata['temp']  # In old SkewT, was sounding.data
    snd_z = sounding.soundingdata['hght']  # In old SkewT, was sounding.data
    dummy_z = []
    dummy_T = []
    if not snd_T.mask[0]: #May cause issue for specific soundings
        dummy_z.append(snd_z[0])
        dummy_T.append(snd_T[0])
        for i, height in enumerate(snd_z):
            if i > 0:
                if snd_z[i] > snd_z[i-1] and not snd_T.mask[i]:
                    dummy_z.append(snd_z[i])
                    dummy_T.append(snd_T[i])
        snd_z = np.array(dummy_z)
        snd_T = np.array(dummy_T)
    return snd_T, snd_z
```

### `add_field_to_radar_object`: transformando array em dado de radar

```python
def add_field_to_radar_object(field, radar, field_name, long_name, standard_name,
                              units='unitless', dz_field='corrected_reflectivity'):
    """
    Adds a newly created field to the Py-ART radar object. If reflectivity is a masked array,
    make the new field masked the same as reflectivity.
    """
    fill_value = -32768
    masked_field = np.ma.asanyarray(field)
    masked_field.mask = masked_field == fill_value
    if hasattr(radar.fields[dz_field]['data'], 'mask'):
        setattr(masked_field, 'mask', 
                np.logical_or(masked_field.mask, radar.fields[dz_field]['data'].mask))
        fill_value = radar.fields[dz_field]['data'].fill_value
    field_dict = {'data': masked_field,
                  'units': units,
                  'long_name': long_name,
                  'standard_name': standard_name,
                  'fill_value': fill_value}
    radar.add_field(field_name, field_dict, replace_existing=True)
    return radar
```

### `radar_coords_to_cart`: Convertendo coordenadas do radar para cartesianas

```python
def radar_coords_to_cart(rng, az, ele, debug=False):

    theta_e = ele * np.pi / 180.0  # elevation angle in radians.
    theta_a = az * np.pi / 180.0  # azimuth angle in radians.
    R = 6371.0 * 1000.0 * 4.0 / 3.0  # effective radius of earth in meters.
    r = rng * 1000.0  # distances to gates in meters.

    z = (r ** 2 + R ** 2 + 2.0 * r * R * np.sin(theta_e)) ** 0.5 - R
    s = R * np.arcsin(r * np.cos(theta_e) / (R + z))  # arc length in m.
    x = s * np.sin(theta_a)
    y = s * np.cos(theta_a)
    return x, y, z
```

Definidas as funções necessárias, o cálculo é feito da seguinte forma:
- Lê os dados da radiossondagem e cria os perfis de altura (`cth_z`) e temperatura (`cth_T`)
- Extrai as variáveis necessárias de `cth`
- Calcula as massas
- Adiciona os campos novos a `cth`

```python
snd = SkewT.Sounding('dados/83779_2017111512Z.txt')
cth_T, cth_z = interpolate_sounding_to_radar(snd, cth)
```

```python
z_corr = cth.fields['corrected_reflectivity']['data']
zdr = cth.fields['differential_reflectivity']['data']
```

```python
mw, mi = csu_liquid_ice_mass.calc_liquid_ice_mass(dbz=z_corr, zdr=zdr, z=cth_z/1000., T=cth_T)
```

```python
cth = add_field_to_radar_object(mw, cth, field_name='MW', units='g m-3',
                                long_name='Liquid Water Mass', standard_name='Liquid Water Mass')
cth = add_field_to_radar_object(mi, cth, field_name='MI', units='g m-3',
                                long_name='Ice Water Mass', standard_name='Ice Water Mass')
```
